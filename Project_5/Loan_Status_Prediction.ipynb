```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn import svm
```

```python
# Data Cleaning and Preprocessing
```

```python
loan_dataset = pd.read_csv('loan.csv')
```

```python
loan_dataset = loan_dataset.dropna()
```

```python
# Label Encoding
```

```python
loan_dataset.head(2)
```

```python
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1},"Gender":{'Male':1,'Female':0},"Married":{'Yes':1,'No':0},"Dependents":{'3+':4},"Education":{'Graduate':1,'Not Graduate':0},"Self_Employed":{'Yes':1,'No':0},"Property_Area":{'Rural':0,'Urban':1,'Semiurban':2}},inplace=True)
```

```python
loan_dataset
```

```python
# Seperting Data and Labels
```

```python
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y = loan_dataset['Loan_Status']
```

```python
X , Y
```

```python
# Train Test Split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2,stratify=Y)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
# Train Model using SVM
```

```python
classifier = svm.SVC(kernel='linear',class_weight='balanced') 
```

```python
# Here we add class_Weight to avoid lazy model(Model predict same value for every prediction EX. Predicting 1 every-time)
```

```python
classifier.fit(X_train,Y_train)
```

```python
# Model Evalution on train data
```

```python
X_train_prediction = classifier.predict(X_train)
print("Train Data Classification Report : \n",classification_report(X_train_prediction,Y_train))
print("Train Data Confusion Metrics : \n",confusion_matrix(X_train_prediction,Y_train))
```

```python
# Model Evalution on test data
```

```python
X_test_prediction = classifier.predict(X_test)
print("Test Data Classification Report : \n",classification_report(X_test_prediction,Y_test))
print("Test Data Confusion Metrics : \n",confusion_matrix(X_test_prediction,Y_test))
```

```python
# Make a predictive System
```

```python
# Use double brackets to keep it as a 2D DataFrame
new_value = X_test.iloc[[25]]

new_prediction = classifier.predict(new_value)
print(new_prediction)

if(new_prediction == 0):
    print("Loan Rejected")
else:
    print("Loan Accepted")
```
